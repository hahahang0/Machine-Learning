```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Simple Linear Equation, y=mx+c

```python
df = pd.read_csv("../placement.csv")
df
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
x = df[["cgpa"]]
y = df["package"]
```

```python
sns.scatterplot(x="cgpa",y="package",data=df)
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# y = mx +c 
```

```python
0.57425647 * 6.89 + -1.02700694
```

```python
lr.predict([[6.89]])
```

```python
y_predict =lr.predict(x)
```

```python
print(y_predict)
```

```python
sns.scatterplot(x="cgpa",y="package",data=df,label="org")
plt.plot(df["cgpa"],y_predict,color="red",label="Predict Line")
plt.legend()
plt.savefig("predict.png",dpi=400)
plt.show()
```
