```python
df["ApplicantIncome"].fillna(df["ApplicantIncome"].mean(),inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
ss.fit(df[["ApplicantIncome"]])
```

```python

df["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(df[["ApplicantIncome"]]),columns=["hello_there"])
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 5))  # Adjust figure size if needed

# Plot 1: Original Data (Before Scaling)
plt.subplot(1, 2, 1)
plt.title("Before Scaling")
sns.histplot(df["ApplicantIncome"], kde=True)  # Use histplot for axes-level plotting
plt.xlabel("ApplicantIncome")

# Plot 2: Scaled Data (After Scaling)
plt.subplot(1, 2, 2)
plt.title("After Scaling")
sns.histplot(df["ApplicantIncome_ss"], kde=True)  # Use histplot for axes-level plotting
plt.xlabel("ApplicantIncome (Scaled)")

# Adjust layout and display
plt.tight_layout()
plt.show()
```
