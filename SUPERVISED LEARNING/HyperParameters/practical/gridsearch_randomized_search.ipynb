```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../dataset/polynomial_salary_data.csv')
```

```python
df.head()
```

```python
x = df[['level']]
y = df["salary"]
```

```python
x,y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test),dt.score(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
temp_dict = {
    "criterion" : ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    "splitter" : ["best","random"],
    "max_depth" : [i for i in range(2,20)]
}
```

```python
gd = GridSearchCV(DecisionTreeRegressor(),param_grid=temp_dict)
```

```python
gd.fit(x_train,y_train)
```

```python
# this is looking for the best param to be used in the algo so , we are inserting this param again into the algorithm to check the best results it can give
gd.best_params_
```

```python
dt = DecisionTreeRegressor(criterion='squared_error', max_depth=8, splitter='best')
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
dt.score(x_train,y_train)
```

```python
gd.best_score_
```

```python
gd.best_estimator_
```

```python
rd = RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=temp_dict,n_iter=20)
```

```python
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
rd.score(x_test,y_test)
```

```python
rd.score(x_train,y_train)
```
