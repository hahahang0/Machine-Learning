```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("../dataset/polynomial_regression_data.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
sns.scatterplot(x='X', y='Y',data=df,hue="X",palette="viridis")
```

```python
x = df[["X"]]
y = df["Y"]
```

```python
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x = pf.transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

<h4>Using linear kernel for polynomial regression dataset</h4>

```python
svm = SVR(kernel="linear")
```

```python
svm.fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
svm.score(x_train,y_train)
```

```python
sns.scatterplot(x='X', y='Y',data=df,hue="X",palette="viridis")
sns.scatterplot(x="X",y=svm.predict(x),data=df)
```

<h4>Using poly kernel for Polynomial regression dataset</h4>

```python
svm1 = SVR(kernel="poly")
```

```python
svm1.fit(x_train,y_train)
```

```python
svm1.score(x_test,y_test)
```

```python
svm1.score(x_train,y_train)
```

```python
sns.scatterplot(x='X', y='Y',data=df,hue="X",palette="viridis")
sns.scatterplot(x="X",y=svm1.predict(x),data=df)
```

<h4>Using rbf</h4>

```python
svm2 = SVR(kernel="rbf")
```

```python
svm2.fit(x_train,y_train)
```
