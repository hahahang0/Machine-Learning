```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("../classification/dataset/placement.csv")
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df["placed"]
```

```python
sns.scatterplot(x="cgpa",y="score",data=df,hue="placed")
```

```python
x 
```

```python
y 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
y_train
```

```python
# sns.kdeplot(data=df["cgpa"])
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
```

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
mnb.score(x_test,y_test),mnb.score(x_train,y_train)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
```

```python
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
bnb.score(x_test,y_test),bnb.score(x_train,y_train)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
```
