```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('../dataset/grocery_transactions.csv')
```

```python
market = []
for i in range(0,df.shape[0]):
    cus = []
    for j in df.columns:
        if type(df[j][i]) == str:
            cus.append(df[j][i])
    market.append(cus)
```

```python
market
```

```python
l = []
for i in market:
    for j in i:
        l.append(j)
```

```python
import collections 
```

```python
p = collections.Counter(l)
```

```python
p 
```

```python
d = {"Item Name:": p.keys(),"values":p.values()}
```

```python
temp_df = pd.DataFrame(d)
```

```python
temp_df.sort_values(by=["values"],ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
```

```python
tr.fit(market)
```

```python
df = pd.DataFrame(tr.transform(market),columns=tr.columns_)
```
