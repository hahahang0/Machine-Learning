```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('../../../GeneralDataset/new_iris.csv')
```

```python
df.head()
```

```python
sns.pairplot(data=df)
```

```python
import scipy.cluster.hierarchy as sch 
```

```python
plt.figure(figsize=(10,20))
sch.dendrogram(sch.linkage(df,method='single',metric="euclidean"))
# plt.savefig("demo.jpg",dpi=400)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage="single")
```

```python
df["predicted"] = ac.fit_predict(df)
```

```python
sns.pairplot(data=df,hue="predicted")
```
