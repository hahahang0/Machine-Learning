```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=500,noise=0.2)
```

```python
y
```

```python
df = {"x1" : x[:,0],"x2" : x[:,1],"y":y}
```

```python
df
```

```python
df  = pd.DataFrame(df)
```

```python
df
```

```python
sns.scatterplot(x="x1",y="x2",hue="y",data=df)
```

```python
x_a = df.iloc[:,:-1]
y_a  = df["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x_a,y_a,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt.score(x_train,y_train),dt.score(x_test,y_test)
```

```python
sv = SVC()
sv.fit(x_train,y_train)
sv.score(x_train,y_train),sv.score(x_test,y_test)
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_train,y_train),sv.score(x_test,y_test)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
li = [("dt1",DecisionTreeClassifier()),("sv1",SVC()),("gnb1",GaussianNB())]
```

```python
vc = VotingClassifier(li,weights=[10,5,5])
```

```python
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train),vc.score(x_test,y_test)
```

```python
prd = {"dt" : dt.predict(x_test),"svm" : sv.predict(x_test),"gnb":gnb.predict(x_test),"vc":vc.predict(x_test)}
```
