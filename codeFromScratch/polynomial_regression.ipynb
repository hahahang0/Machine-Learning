```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('dataset.csv')
# type(df)
```

```python
df.head()
```

```python
df.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[["X"]],df[['y']],random_state=42,test_size=0.1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
plt.figure(figsize=(20,10))
plt.scatter(df['X'],df['y'])
plt.xlabel("X")
plt.ylabel("y")
plt.tight_layout()
plt.show()
```

```python
### lets try linear regression and try to plot the graph and look.
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
X_train
```

```python
y_pred = lr.predict(X_train)
```

```python
y_pred.shape
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X_train, y_train, label="Training data")            
plt.plot(X_train, y_pred, c='red', linewidth=3, label="Model") 
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# not valid , so we use polynomial regression , example using sklearn and then from scratch 
```

```python
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
lr1 = LinearRegression()
lr1.fit(X_train_poly,y_train)
```

```python
X_train_sorted = np.sort(X_train, axis=0)
X_sorted_poly = poly.transform(X_train_sorted)
y_pred_curve = lr1.predict(X_sorted_poly)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X_train, y_train, label="Training data")  
plt.plot(X_train_sorted,y_pred_curve, c='red', linewidth=3, label="Model") 
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.tight_layout()
plt.show()
```
