```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("../dataset/synthetic_ice_cream_sales.csv")
```

```python
df.sample(10)
```

```python
sns.scatterplot(x="Temperature",y="Ice Cream Sales",data=df,hue="Ice Cream Sales",palette="coolwarm")
```

```python
x = df[["Temperature"]]
y = df[["Ice Cream Sales"]]
```

```python
x,y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

<h4>Using Linear Regression</h4>

```python
svm = SVR(kernel="linear")
```

```python
svm.fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
svm.score(x_train,y_train)
```

```python
sns.scatterplot(x="Temperature",y="Ice Cream Sales",data=df,hue="Ice Cream Sales",palette="coolwarm")
sns.scatterplot(x=df["Temperature"],y=svm.predict(x),color="green")
plt.show()
```

<h4>Using Polynomial</h4>

```python
svm1 = SVR(kernel="poly")
svm1.fit(x_train,y_train)
```

```python
svm1.score(x_test,y_test)
```

```python
svm1.score(x_train,y_train)
```

```python
sns.scatterplot(x="Temperature",y="Ice Cream Sales",data=df,hue="Ice Cream Sales",palette="coolwarm")
sns.scatterplot(x=df["Temperature"],y=svm1.predict(x),color="green")
plt.show()
```
