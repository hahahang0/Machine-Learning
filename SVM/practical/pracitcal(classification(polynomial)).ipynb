```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC,SVR
from mlxtend.plotting import plot_decision_regions

```

```python
df = pd.read_csv("../dataset/polynomial_classification_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.scatterplot(
    x="Feature1",
    y="Feature2",
    hue="Class",  # Color by temperature
    # Adjust the size range
    data=df,
    palette="viridis"  # Color palette
)
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df["Class"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
svm = SVC(kernel="rbf",C=1,gamma=2)
svm2 = SVC(kernel="poly")
```

```python
svm.fit(x_train,y_train)
svm2.fit(x_train,y_train)
```

```python
print(svm.score(x_test,y_test))
print(svm2.score(x_test,y_test))
```

```python
print(svm.score(x_train,y_train))
print(svm2.score(x_train,y_train))
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svm,legend=2)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svm2,legend=2)
```
