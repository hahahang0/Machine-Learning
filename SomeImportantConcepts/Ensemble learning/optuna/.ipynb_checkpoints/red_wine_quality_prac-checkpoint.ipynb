```python
class Graphs:
    def __init__(self,df):
        self.df = df 
    def kdeplot(self):
         numeric_cols = self.df.select_dtypes(include='number').columns
         for col in numeric_cols:
            plt.figure(figsize=(20, 10))
            sns.kdeplot(data=self.df, x=col, fill=True)
            plt.title(f'KDE Plot of "{col}"')
            plt.xlabel(col)
            plt.ylabel('Density')
            plt.grid(True)
            plt.tight_layout()
            plt.show()
        
    def boxplot(self):
         numeric_cols = self.df.select_dtypes(include='number').columns
         for col in numeric_cols:
            plt.figure(figsize=(20, 10))
            sns.boxplot(data=self.df, x=col, fill=True)
            plt.title(f'KDE Plot of "{col}"')
            plt.xlabel(col)
            plt.ylabel('Density')
            plt.grid(True)
            plt.tight_layout()
            plt.show()
    def violinplot(self):
         numeric_cols = self.df.select_dtypes(include='number').columns
         for col in numeric_cols:
            plt.figure(figsize=(20, 10))
            sns.violinplot(data=self.df, x=col, fill=True)
            plt.title(f'KDE Plot of "{col}"')
            plt.xlabel(col)
            plt.ylabel('Density')
            plt.grid(True)
            plt.tight_layout()
            plt.show()
    
```

```python
def removing_outliers_iqr(df):
    df_copy = df.copy()
    
    for column in df_copy.columns:
        q1 = df_copy[column].quantile(0.25)
        q3 = df_copy[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1-1.5*iqr
        upper_bound = q3+1.5*iqr
        df_copy = df_copy[(df_copy[column] >= lower_bound) & (df_copy[column] <= upper_bound)]
    return df_copy
```

```python
filtered_df = removing_outliers_iqr(df)
```
